# Euclidean and Non-Euclidean Geometry

We simulate the parallel postulate in Euclidean and Non-Euclidean (spherical) Spaces. The postulate states that given a line and a point not on the line, there is exactly one line parallel to the given line that passes through the given point. In Euclidean space, the parallel line is unique, while in Non-Euclidean space, there are no parallel lines.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up figure with two subplots
fig = plt.figure(figsize=(12, 5))

# --- Euclidean Space Visualization ---
ax1 = fig.add_subplot(121, projection='3d')
ax1.set_title("Euclidean Space (Flat Plane)")

# Define a flat plane (z = 0 for simplicity)
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)
Z = np.zeros_like(X)

# Plot the plane
ax1.plot_surface(X, Y, Z, alpha=0.3, color='lightblue')

# Reference line on the plane (y = 0, z = 0)
x_ref = np.linspace(-2, 2, 100)
y_ref = np.zeros_like(x_ref)
z_ref = np.zeros_like(x_ref)
ax1.plot(x_ref, y_ref, z_ref, color='red', label='Reference Line')

# Point not on the reference line (e.g., at x=0, y=1)
point = [0, 1, 0]
ax1.scatter(point[0], point[1], point[2], color='black', label='Point P')

# Parallel line through the point (y = 1, z = 0)
x_parallel = np.linspace(-2, 2, 100)
y_parallel = np.ones_like(x_parallel)
z_parallel = np.zeros_like(x_parallel)
ax1.plot(x_parallel, y_parallel, z_parallel, color='green', label='Parallel Line')

ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.legend()

# --- Spherical Space Visualization ---
ax2 = fig.add_subplot(122, projection='3d')
ax2.set_title("Spherical Space")

# Define a sphere
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
X_sphere = np.outer(np.cos(u), np.sin(v))
Y_sphere = np.outer(np.sin(u), np.sin(v))
Z_sphere = np.outer(np.ones(np.size(u)), np.cos(v))

# Plot the sphere
ax2.plot_surface(X_sphere, Y_sphere, Z_sphere, alpha=0.3, color='lightblue')

# Reference great circle (equator, z = 0 plane intersection)
theta = np.linspace(0, 2 * np.pi, 100)
x_ref_sphere = np.cos(theta)
y_ref_sphere = np.sin(theta)
z_ref_sphere = np.zeros_like(theta)
ax2.plot(x_ref_sphere, y_ref_sphere, z_ref_sphere, color='red', label='Reference Great Circle')

# Point not on the reference line (e.g., north pole-ish, shifted)
point_sphere = [0, 0, 1]
ax2.scatter(point_sphere[0], point_sphere[1], point_sphere[2], color='black', label='Point P')

# Great circle through the point (e.g., a meridian passing through the point)
phi = np.linspace(0, 2 * np.pi, 100)
x_parallel_sphere = np.sin(phi) * np.cos(np.pi / 4)  # Tilted great circle
y_parallel_sphere = np.sin(phi) * np.sin(np.pi / 4)
z_parallel_sphere = np.cos(phi)
ax2.plot(x_parallel_sphere, y_parallel_sphere, z_parallel_sphere, color='green', label='Great Circle through P')

ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.legend()

# Adjust layout and display
plt.tight_layout()
plt.show()

# Quick simulation explanation in comments
print("""
Euclidean Space:
- The red line is the reference line.
- The green line is parallel to it through point P.
- Only one such parallel exists, satisfying Euclid's fifth postulate.

Spherical Space:
- The red curve is a great circle (like the equator).
- The green curve is another great circle through point P.
- They intersect (e.g., at poles), showing no parallels exist in spherical geometry.
""")
```
